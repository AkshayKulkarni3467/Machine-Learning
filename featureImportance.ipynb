```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Datasets/mnist_train.csv')
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
sns.heatmap(X.iloc[5].values.reshape(28,28))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X,y)
```

```python
#Features in the center of the matrix are important to guess the numbers
sns.heatmap(rf.feature_importances_.reshape(28,28))
```
